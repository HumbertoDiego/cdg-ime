### Avaliações

**1ª VE oral:** apresentação curta, de 8 a 10 minutos, sobre o estudo inicial do projeto escolhido. O aluno deverá demonstrar que compreendeu o problema, delimitou um objetivo viável, identificou dados e ferramentas disponíveis, levantou referências iniciais e propôs um plano de trabalho compatível com o semestre.

Sugestão:
- tema escolhido e justificativa;
- problema de pesquisa;
- objetivo geral refinado;
- objetivos específicos;
- dados necessários e possíveis fontes;
- ferramentas previstas, como Python, QGIS, GeoPandas, GDAL, APIs, STAC, LLM local, drone ou outras;
- metodologia preliminar;
- produto esperado, como mapa, aplicação, relatório, modelo, comparação técnica ou fluxo automatizado;
- principais riscos/dificuldades;
- cronograma inicial para as próximas semanas.

**2ª VE escrita:** Sobre os assuntos ministrados em aula.

**3ª VE oral:** apresentação mais longa, de 20 a 30 minutos, sobre o estudo parcial do projeto escolhido. O aluno deverá demonstrar que já executou parte substancial da metodologia proposta, apresentando os dados efetivamente utilizados, os procedimentos de processamento e análise realizados, os resultados parciais obtidos, as dificuldades encontradas, as decisões metodológicas tomadas e o plano objetivo para concluir o produto final.

Itens mínimos:
- problema, objetivo geral e objetivos específicos atualizados;
- dados utilizados, fontes, critérios de seleção e limitações;
- fluxo metodológico já implementado;
- scripts, notebooks, modelos, mapas ou aplicações em desenvolvimento;
- resultados parciais com figuras, tabelas, mapas ou métricas;
- avaliação preliminar da qualidade dos dados e dos produtos;
- comparação entre o planejado e o executado;
- principais dificuldades técnicas e como foram tratadas;
- próximos passos até a entrega final;
- riscos restantes e plano de mitigação.
